```python
from sklearn.preprocessing import OneHotEncoder

data = pd.DataFrame({
    "Gender": ["Male", "Female", "Male", "Female"],
    "Sleep Duration": ["7-8 hours", "5-6 hours", "More than 8 hours", "Less than 5 hours"],
    "Dietary Habits": ["Moderate", "Unhealthy", "Healthy", "Moderate"],
    "Have you ever had suicidal thoughts ?": ["Yes", "No", "No", "Yes"],
    "Family History of Mental Illness": ["Yes", "No", "Yes", "No"],
   
})


ohe = OneHotEncoder(sparse_output=False)


encoded = ohe.fit_transform(data[["Gender", "Sleep Duration", "Dietary Habits", "Have you ever had suicidal thoughts ?", "Family History of Mental Illness"]])


one_hot_encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(["Gender", "Sleep Duration", "Dietary Habits", "Have you ever had suicidal thoughts ?", "Family History of Mental Illness",]))

print(df)
```

```python
one_hot_encoded_df.head()
```

```python
one_hot_encoded_df.corr()
```

```python
corr = one_hot_encoded_df.corr(numeric_only=True)

mask = np.triu(np.ones_like(corr, dtype=bool))
plt.figure(figsize=(12, 8))
sns.heatmap(
    corr,
    mask=mask,
    cmap="coolwarm",
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    square=True,
    cbar_kws={"shrink": 0.7},
)

plt.title("Correlation Heatmap", fontsize=16, fontweight="bold")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
data = pd.DataFrame({
    "Gender": ["Male", "Female", "Male", "Female"],
    "Sleep Duration": ["7-8 hours", "5-6 hours", "More than 8 hours", "Less than 5 hours"],
    "Dietary Habits": ["Moderate", "Unhealthy", "Healthy", "Moderate"],
    "Have you ever had suicidal thoughts ?": ["Yes", "No", "No", "Yes"],
    "Family History of Mental Illness": ["Yes", "No", "Yes", "No"]
})

le = LabelEncoder()

LabelEncoder_data = data[["Gender", "Sleep Duration", "Dietary Habits", "Have you ever had suicidal thoughts ?", "Family History of Mental Illness"]].apply(le.fit_transform)
```

```python
corr = LabelEncoder_data.corr(numeric_only=True)

mask = np.triu(np.ones_like(corr, dtype=bool))
plt.figure(figsize=(12, 8))
sns.heatmap(
    corr,
    mask=mask,
    cmap="coolwarm",
    annot=True,
    fmt=".2f",
    linewidths=0.5,
    square=True,
    cbar_kws={"shrink": 0.7},
)

plt.title("Correlation Heatmap", fontsize=16, fontweight="bold")
plt.show()
```
