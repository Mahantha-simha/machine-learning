```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

np.set_printoptions(precision=2)
```

```python
# make 4-class dataset for classification
classes = 4
m = 100
centers = [[-5, 2], [-2, -2], [1, 2], [5, -2]]
std = 1.0
X_train, y_train = make_blobs(n_samples=m, centers=centers, cluster_std=std,random_state=30)

```

```python
# Define the plt_mc function
def plt_mc(X, y, classes, centers, std):
	plt.figure(figsize=(8, 6))
	for class_value in range(classes):
		plt.scatter(X[y == class_value, 0], X[y == class_value, 1], label=f'Class {class_value}')
	for center in centers:
		plt.scatter(center[0], center[1], s=200, c='black', marker='x', label='Center')
	plt.title(f'Multi-Class Data with {classes} Classes')
	plt.xlabel('Feature 1')
	plt.ylabel('Feature 2')
	plt.legend()
	plt.grid(True)
	plt.show()

# Call the plt_mc function
plt_mc(X_train, y_train, classes, centers, std=std)
```
