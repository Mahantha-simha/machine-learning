```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('Global_Cybersecurity_Threats_2015-2024.csv')
df.info()
```

```python
sns.histplot(data=df, x='Financial Loss (in Million $)', bins=20)
plt.title('Distribution of Cybersecurity Threats Over Years')
plt.xlabel('Financial Loss (in Million $)')
plt.ylabel('Count')
```
